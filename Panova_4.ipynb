```python
import numpy as np
import math as m
```

```python
N = 20
m = np.zeros((N,N)) # инициализация матриц
v = np.zeros((N,1))
x = np.zeros ((N,1))
for i in range(1,19): # заполнение коэффициентов
    m[i][i-1] = 1
    m[i][i] = -2
    m[i][i+1] = 1
    v[i][0] = 2/(i*i)
for i in range(19):
    m[19][i] = 2
m[0][0] = 1
m[19][0] = 1
m[19][19] = 1
v[0][0] = 1
v[19][0] = -N/3
print(m)
print()
print(v)
```

```python
w,a = np.linalg.eig(m)
#print (w)
min = 0
max = 0
for i in range(len(m)):
    if min > w[i]:
        min = w[i]
    if max < w[i]:
        max = w[i]
print('Собственные значения матрицы ')
print('минимальное ', min)
print('максимальное ', max)
```

```python
cond = np.linalg.cond(m)
print('число обусловленности матрицы ', cond)
```

Реализация метода Гаусса

```python
for k in range(0,N): # прямой ход
    t = m[k][k]
    for j in range(k,N):
        m[k][j] = m[k][j]/t
    v[k] = v[k]/t
    for i in range(k+1,N):
        t = m[i][k]
        for j in range(k,N):
            m[i][j] = m[i][j] - m[k][j]*t
        v[i] = v[i] - v[k]*t
        
for k in reversed(range(N)): # обратный ход
    x[k] = v[k]
    for j in range(k+1,N):
        x[k] = x[k]-m[k][j]*x[j]
print(x)
```

Вектор невязкости Метода Гаусса

```python
print(v - np.dot(m, x))
```

Реализация метода Якоби

```python
x0 = np.zeros ((N,1))
x1 = np.zeros ((N,1))
E = 0.0001
max = 1

for i in range(N): # преобразует матрицу 
    t = m[i][i]
    for j in range(N):
        m[i][j] = -m[i][j]/t
    m[i][i] = 0
    v[i] = v[i]/t
#print(m)
    
for i in range(N): # задается начальное приближение
    x0[i] = v[i]

k = 1 # первая итерация 
while (max >= E):
    for i in range (N): # вычисление x^(k)
        x1[i] = v[i]
        for j in range(N):
            x1[i] = x1[i] + m[i][j]*x0[j]

    max = 0 # вичисление отклонения
    for i in range(N):
        if abs(x1[i] - x0[i]) > max:
            max = abs(x1[i] - x0[i])

    k = k + 1 # переход на следующую итерацию

    for i in range(N):
        x0[i] = x1[i]

print(x0)
```

Вектор невязкости Метода Якоби

```python
print(v - np.dot(m, x0))
```

Метод Якоби построчно

```python
E = 0.000001
max = 1
N = 20
f = np.zeros(N)
x = np.zeros(N)
x0 = np.zeros(N) # предыдущие x. начальное приближение равно вектору f

for i in range(20):
    x0[i] = 10
while (max >= E):
    for k in range(N): # счетчик строки
        a = np.zeros(N)
        if k == 0:
            a[0] = 1
            f[0] = 1
        elif k == 19:
            a[0] = 1
            a[19] = 1
            f[19] = -19/3
            for i in range(1,19):
                a[i] = 2
        else:
            a[k] = -2
            a[k-1] = 1
            a[k+1] = 1
            f[k] = 2/k/k
       
        s = 0    
        for i in range(N):
            if i != k:
                s += a[i]*x0[i]
        x[k] = (f[k] - s)/a[k]

        max = 0 # вичисление отклонения
        for i in range(N):
            if abs(x[i] - x0[i]) > max:
                max = abs(x[i] - x0[i])
            x0[i] = x[i]
print(x)
```
