```python
"""
@author: Panova Maria 428

    arctg(x)   . 
        .
   .

"""

import numpy as np
import math as m
import matplotlib.pyplot as plt

N = 2000
h = 0.01  #
x0 = -10

X = np.zeros(N) 
py = np.zeros(N)  #       
ty = np.zeros(N)  #    ,  
dy = np.zeros(N)  #     

x = x0
for i in range(N):  #  
    x += h
    y = (m.atan(x+h)-2*m.atan(x)+m.atan(x-h))/(h*h) #     
    X[i] = x
    py[i] = y

x = x0    
for i in range(N):  # 
    x += h
    y = (-2*x)/(x*x*x*x+2*x*x+1) #  ,  
    X[i] = x
    ty[i] = y
    
x = x0
for i in range(N): #  
    x += h
    y = py[i]-ty[i]
    X[i] = x
    dy[i] = y

plt.figure()   #  
plt.plot (X, py)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(True)
plt.title ('$central-difference-formula$')
plt.show()

plt.figure()   #  
plt.plot (X, ty)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(True)
plt.title ('$manually-calculator-formula$')
plt.show()

plt.figure()   #   
plt.plot (X, dy)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(True)
plt.title ('$difference-error-graph$')
plt.show()

```
