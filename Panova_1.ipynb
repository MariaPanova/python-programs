```python
"""
@author: Panova Maria 428

Вычислить вторую производную функции arctg(x) методом центральной разности. 
Построить график производной и сравнить с графиком аналитической производной.
Построить разностный график ошибок.

"""

import numpy as np
import math as m
import matplotlib.pyplot as plt

N = 2000
h = 0.01  #шаг
x0 = -10

X = np.zeros(N) 
py = np.zeros(N)  #массив для значений второй производной методом центральной разности
ty = np.zeros(N)  #массив для значений второй производной, найденной вручную
dy = np.zeros(N)  #массив для разностного графика ошибок 

x = x0
for i in range(N):  #метод центральной разности
    x += h
    y = (m.atan(x+h)-2*m.atan(x)+m.atan(x-h))/(h*h) #формула второй производной методом центральной разности
    X[i] = x
    py[i] = y

x = x0    
for i in range(N):  #метод аналитический
    x += h
    y = (-2*x)/(x*x*x*x+2*x*x+1) #формула второй производной, найдена вручную
    X[i] = x
    ty[i] = y
    
x = x0
for i in range(N): #разностный график ошибок
    x += h
    y = py[i]-ty[i]
    X[i] = x
    dy[i] = y

plt.figure()   #построение первого графика
plt.plot (X, py)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(True)
plt.title ('$central-difference-formula$')
plt.show()

plt.figure()   #построение второго графика
plt.plot (X, ty)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(True)
plt.title ('$manually-calculator-formula$')
plt.show()

plt.figure()   #построение разностного графика ошибок
plt.plot (X, dy)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(True)
plt.title ('$difference-error-graph$')
plt.show()

```
